```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"# load dataset into Pandas DataFrame
df = pd.read_csv(url, names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
global grafico #figure
#Function scatter_plot group data by argument name, plot and edit labels
def scatter_plot(x_label,y_label,z_label,clase,c,m,label):
    x =df[df['target'] == clase ][x_label] #groupby Name column x_label
    y = df[ df['target'] == clase ][y_label]
    z = df[ df['target'] == clase ][z_label]
    # s: size point; alpha: transparent 0, opaque 1; label:legend
    grafico.scatter(x,y,z,color=c, edgecolors='k',s=50, alpha=0.9, marker=m,label=label)
    grafico.set_xlabel(x_label)
    grafico.set_ylabel(y_label)
    grafico.set_zlabel(z_label)
    return 

grafico = plt.figure().gca(projection='3d')  #new figure
scatter_plot('sepal length','sepal width','petal length','Iris-virginica','g','o','Iris-virginica')
scatter_plot('sepal length','sepal width','petal length','Iris-versicolor','b','o','Iris-versicolor')
scatter_plot('sepal length','sepal width','petal length','Iris-setosa','r','o','Iris-setosa')
plt.legend()
plt.show()
```
