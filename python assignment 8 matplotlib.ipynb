```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
```

```python
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [1, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
data=pd.DataFrame({"Max":Max,"Min":Min})
data
```

```python
plt.plot([i for i in data.index],data['Max'],"bo",data["Min"],"ro")
```

```python
from scipy import optimize

def yearly_temps(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, [i for i in data.index], data['Max'], [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, [i for i in data.index], data["Min"], [-40, 20, 0])
```

```python
days = np.linspace(0, 12, num=365)

plt.figure(figsize=(8,6))
plt.plot([i for i in data.index], data["Max"], 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot([i for i in data.index], data['Min'], 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
df2=titanic[['fare',"age","sex"]].dropna(axis=0)
sex=[j.replace("male","2") for j in [i.replace("female","1") for i in df2["sex"]]]
df2[['sex']]=sex
sex=df2['sex'].astype(int)
plt.scatter(x=df2["fare"],y=df2["age"],c=sex,cmap="winter")
cbar = plt.colorbar()
cbar.set_label('Sex')
```
